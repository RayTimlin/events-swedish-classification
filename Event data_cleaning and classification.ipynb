# Utföra analys på  eventdata

### Load the modules
__CSV__ fil ska laddas in. De första 5 incidenter ska synas.


```python
import numpy as np
import pandas as pd

colnames=['cityID', 'Time', 'Name', 'Description', 'Amount', 'Kategori'] 

# Hämta in excel
df = pd.read_csv('csv/event_data.csv', names=colnames, delimiter=',', encoding='utf-8')

df.head()
```

### Creating a subset to look at closer

Just working with the description and the kategori sections

```python
df_analyse = df[['Description', 'Kategori']]

#show entire value not just a trunctuated version 
pd.set_option('display.max_colwidth', -1)

print(df_analyse)
```

### Get rid of the html
Just extract the text information

```python
from bs4 import BeautifulSoup

# Converting to string to avoid the float problem
df_analyse['Description']  = df_analyse['Description'] .astype(str)

# Extracting the text
df_analyse = df_analyse.applymap(lambda text: BeautifulSoup(text, 'html.parser').get_text())

df_analyse.head(20)
```

###  Create a new column which captures the amount of words in each event.

```python
wordCount = 5

# Create a new column with the number of chars in the question
df_analyse['wordCount']=df_analyse['Description'].str.split().str.len()

print ('There are ', len(df_analyse[df_analyse['wordCount']<= wordCount]), ' different occurances with a word count <= ', wordCount)

df_analyse[df_analyse['wordCount']==wordCount].head(50)

```

### Get rid of all sentences below the defined wordcount

```python
df_analyse=df_analyse[df_analyse['wordCount']>= wordCount]  
```

### Retain only the top categories.
Obs - Alter the categoryNo variable to retain a certain number of categories


```python
categoryNo = 30

print ('There are ', len(df_analyse['Kategori'].unique()), 'different categories')
# Lets tryout the Pandas bar charts, read more here https://pandas.pydata.org/pandas-docs/stable/user_guide/style.html#Bar-charts
df_analyse['Kategori'].value_counts().head(categoryNo).to_frame().style.bar()
```

### Get rid of everything that has less than 1000 occurances

```python
df_analyse=df_analyse[df_analyse.groupby('Kategori')['Kategori'].transform('count')>=1000]
df_analyse['Kategori'].value_counts().head(50).to_frame().style.bar()
```

### Fixing some Swedish text



```python
df_analyse['Description'] = df_analyse['Description'].str.replace('&aring;', 'å')
df_analyse['Description'] = df_analyse['Description'].str.replace('&ouml;', 'ö')
df_analyse['Description'] = df_analyse['Description'].str.replace('&auml;', 'å')
df_analyse['Description'] = df_analyse['Description'].str.replace('&Ouml;', 'Ö')
df_analyse['Description'] = df_analyse['Description'].str.replace('&Auml;', 'Ä')
df_analyse['Description'] = df_analyse['Description'].str.replace('&Aring;', 'Å')

```

### Convert into a pivot table to get data to sit right way

```python
df_pivot = pd.pivot_table(df_analyse,index=["Description"])

df_analyse.head()
```

### Convert float data to Int and roundup

Also checking out how the data looks

```python
colnames = list(df_pivot.columns)

# Not useful anymore. This list should be used later
colnames.remove('wordCount')

# Index not included amongst the column names?
print(colnames)

for i in colnames:
    df_pivot[i]  = df_pivot[i].apply(np.ceil).astype(int)

df_pivot.head(10)
    
    
```

# Part 2 - Classifying the Swedish text into different categories

Importing in some modules and reset the index so its not description anymore.

```python
from simpletransformers.classification import MultiLabelClassificationModel
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split


df_pivot.reset_index(level=0, inplace=True)
df_pivot = shuffle(df_pivot)

print("The number of rows and columns in the dataset is: {}".format(df_pivot.shape))
```

### Convert relevant variables into one_hot_list

```python
# Convert these into a single list. FIX so these don't need to be hardcoded
df_pivot['one_hot_labels'] = df_pivot[['Kategori_fest', 'Kategori_fika', 'Kategori_film', 'Kategori_gratis', 'Kategori_konsert', 'Kategori_kultur', 'Kategori_mat', 'Kategori_musik', 'Kategori_promenad', 'Kategori_pub', 'Kategori_teater']].values.tolist()

# Create a new df 
df_test = df_pivot[['Description','one_hot_labels']]

df_test.head()
```

### Split the data into train and test


```python
# print (df_test.shape)

# # Get 25% for validation
df_validation = df_test.sample(frac=0.25, random_state=0)
df_test = df_test.drop(df_validation.index)



train_df, eval_df = train_test_split(df_test, test_size = 0.2)

print ('The size of training dataset is: ', train_df.shape)
print ('The size of test dataset is: ', eval_df.shape)
print ('The size of validation dataset is: ', df_validation.shape)
```

### Run the model - this takes a long time to run - 12 hours?

```python
# If running into CPU problems then alter the below (to lower values)
batchSize = 4
sequenceLength = 256

#How many times the model should be run to learn on
epochNo = 3


model = MultiLabelClassificationModel("bert", "KB/bert-base-swedish-cased", use_cuda=False, num_labels=11, args={'train_batch_size':batchSize, 'gradient_accumulation_steps':16, 'learning_rate': 3e-5, 'num_train_epochs': epochNo, 'fp16': False, 'overwrite_output_dir':True, 'max_seq_length': sequenceLength })
model.train_model(train_df, eval_df=eval_df)
```

### Evaluate the model 

```python
result, model_outputs, wrong_predictions = model.eval_model(eval_df)

print(result)
print(model_outputs)
```
